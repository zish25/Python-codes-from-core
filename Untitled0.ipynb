```python
## HW 5 programs related to coditional statement
## 1. Check if a number is positive, negative, or zero
num = int(input("Enter a number: "))
if num > 0:
    print("The number is positive.")
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

```python
## 2. Check if a number is even or odd
num = int(input("Enter a number: ")) # Get input from the user and convert it to an integer
if num % 2 == 0: # Check if the number is divisible by 2 with no remainder
    print("The number is even.") # If the remainder is 0, the number is even
else: # Otherwise, the number is odd
    print("The number is odd.")
```

```python
### 3. Find the largest of three numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print(a, "is the largest number.")
elif b >= a and b >= c:
    print(b, "is the largest number.")
else:
    print(c, "is the largest number.")
```

```python
### 4.Character classification (vowel or consonant)
ch = input("Enter a character: ")

if ch.lower() in ('a', 'e', 'i', 'o', 'u'):
    print("It is a vowel.")
else:
    print("It is a consonant.")
```

```python
## relational operators :- used to compare two values and return a Boolean result (true or false).
## operators :- ==,!=,>,<,>=,<=
x = 10
y = 20
print(x > y)   # False
print(x != y)  # True
```

```python
## Logical operators :- used to combine multiple conditions (and , or , not)
a = 10
b = 5
print(a > 0 and b > 0)  # True
print(a > 0 or b < 0)   # True
print(not (a > b))      # False
```

```python
#Boolean Operators :- Boolean represents a value that is either True or False.
x = 5
y = 10
result = (x < y)
print(result)        # True
print(type(result))  # <class 'bool'>
```
